```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('sakila.db')
```

```python
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
    
```

```python
#query uses a syntax specific to sqlite, and its purpose is to select all the tables in a database

query = '''
    SELECT *
    FROM sqlite_master
    WHERE type = 'table';
'''

tables = sql_to_df(query)
tables
```

```python
query = '''
    SELECT first_name, last_name
    FROM customer
    ORDER BY last_name DESC
'''

customer_names = sql_to_df(query)
customer_names
```

```python
#head() is a pandas dataFrame method that will show the "head" of the customer_names DataFrame. What that means is that it will return only the first few records

print(customer_names.head())
```
